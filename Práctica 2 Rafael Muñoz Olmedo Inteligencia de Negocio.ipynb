# Normalización

```python
def norm_to_zero_one(df):
    return (df - df.min()) * 1.0 / (df.max() - df.min())
```

```python
df[usadas] = norm_to_zero_one(df[usadas])
```

# Creación de nuestros modelos

```python
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.cluster import Birch
from sklearn.cluster import MeanShift
from sklearn.cluster import AgglomerativeClustering
```

```python
def kmeans_model():
    return KMeans(init='k-means++', n_clusters=4, n_init=3)
```

```python
def dbscan_model(): 
    return DBSCAN(min_samples=2) 
```

```python
def birch_model():
    return Birch()
```

```python
def meanshift_model():
    return MeanShift()
```

```python
def agglomerative_model():
    return AgglomerativeClustering()
```

```python
def seleccion_modelo(modelo):
    if modelo== 'kmeans':
        return kmeans_model()
    elif modelo== 'dbscan':
        return dbscan_model()
    elif modelo== 'birch':
        return birch_model()
    elif modelo== 'meanshift':
        return meanshift_model()
    elif modelo== 'agglomerative':
        return agglomerative_model()
    
        
```

# Obtención de métricas

```python
import time
from sklearn import metrics
from math import floor
```

```python
def metricas(df, model):
   
    t = time.time()
    cluster_predict = model.fit_predict(df)
    tiempo = time.time() -t
   
    metric_CH = metrics.calinski_harabasz_score(df, cluster_predict)
  
    muestra_silhoutte = 0.2 if (len(df) > 10000) else 1.0
    metric_SC = metrics.silhouette_score(df, cluster_predict, metric='euclidean', sample_size=floor(muestra_silhoutte*len(df)), random_state=123456)
    
    
    return tiempo, metric_CH, metric_SC, cluster_predict
```

```python
def imprimir_metricas (time, metric_CH, metric_SC):
    print("Tiempo en segundos: ", time)
    print("Calinski Harabasz: ", metric_CH)
    print("Silhoutte score: ", metric_SC)
```

# Gráficas

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
def df_clusters(df, cluster_predict):
    clusters = pd.DataFrame(cluster_predict,index=df.index,columns=['cluster'])
    return clusters
```

```python
def centers(df, modelo):
    centers = pd.DataFrame(modelo.cluster_centers_,columns=list(df))
    centers_desnormal = centers.copy()
    #se convierten los centros a los rangos originales antes de normalizar
    for var in list(centers):
        centers_desnormal[var] = df[var].min() + centers[var] * (df[var].max() - df[var].min())
    return centers, centers_desnormal
```

```python
def imprimir_heatmap(centers):
    ax = sns.heatmap(centers, linewidth=0.5)
    plt.show()
```

```python
def imprimir_scatter_matrix(df, clusters):
    X = pd.concat([df, clusters], axis=1)
    sns.set()
    variables = list(X)
    variables.remove('cluster')
    sns_plot = sns.pairplot(X, vars=variables, hue="cluster", palette='Paired', plot_kws={"s": 25}, diag_kind="hist") #en hue indicamos que la columna 'cluster' define los colores
    sns_plot.fig.subplots_adjust(wspace=.03, hspace=.03);
    sns_plot.savefig("scattermatrix.png")
    print("")

    
```

# Construcción de casos 

```python
def caso(df, modelo):
    modelo = seleccion_modelo(modelo)
    time, metric_CH, metric_SC, cluster_predict = metricas(df, modelo)
    imprimir_metricas (time, metric_CH, metric_SC)
    clusters =  df_clusters(df, cluster_predict)
    centros, centers_desnormal = centers(df, modelo)
    imprimir_heatmap(centers_desnormal)
    imprimir_scatter_matrix(df, clusters)
```

```python
def caso_db(df, modelo):
    modelo = seleccion_modelo(modelo)
    time, metric_CH, metric_SC, cluster_predict = metricas(df, modelo)
    imprimir_metricas (time, metric_CH, metric_SC)
    clusters =  df_clusters(df, cluster_predict)
    imprimir_scatter_matrix(df, clusters)
```
